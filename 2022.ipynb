#인구피라미드

```python
#남자데이터정의
import pandas as pd #라이브러리 불러오기
df_m = pd.read_excel('/content/drive/MyDrive/202208_연령별인구현황_월간.xlsx', skiprows = 3, index_col='행정기관', usecols='B,E:Y') #현황별 인구수 파일 불러오기, '행정기관'을 인데스로 설정
df_m

df_m.iloc[0]

#인구수자료형에 , 가 있으면 문자로 인식되어서 , 를 공백으로 변환 ->int자료형 인식할 수 있음
df_m.iloc[0] = df_m.iloc[0].str.replace(',','').astype(int)
df_m

#여자데이터정의
import pandas as pd #라이브러리 불러오기
df_w = pd.read_excel('/content/drive/MyDrive/202208_연령별인구현황_월간.xlsx', skiprows = 3, index_col='행정기관', usecols='B,AB:AV') #현황별 인구수 파일 불러오기, '행정기관'을 인데스로 설정, B, E~Y(남자)를 열 불러오기
df_w

df_w.columns = df_m.columns #컬럼명 통일
df_w.columns #위 데이터플레임에서 0~4세.1이 0~4세로 통일 된 것을 확인할 수 있음

#인구수자료형에 , 가 있으면 문자로 인식되어서 , 를 공백으로 변환 ->int자료형 인식할 수 있음
df_w.iloc[0] = df_w.iloc[0].str.replace(',','').astype(int)

df_w

#데이터시각화

import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' #window
#matplotlib.rcParams['font.family'] ='AppleGothic' #Mac
matplotlib.rcParams['font.size'] = 15  #글자크기
matplotlib.rcParams['axes.unicode_minus'] = False #한글 폰트 사용시, 마이너스 글자가 깨지는 현상을 해결

plt.figure(figsize=(10,7)) # 그래프 크기 키우기
plt.barh(df_m.columns, -df_m.iloc[0]/1000)#barh 는 가로 막대그래프 /1000 함으로써 단위 : 1000(명) -는 남자데이터를 열기준 왼쪽에 표시하기 위함
plt.barh(df_w.columns, df_w.iloc[0]/1000)#barh 는 가로 막대그래프 /1000 함으로써 단위 : 1000(명)
plt.title('2022 대한민국 인구 피라미드') #그래프에 제목 추가
plt.savefig('/content/drive/MyDrive/2022_인구피라미드.png', dpi = 100) #파일로 저장하기
plt.show()
```
